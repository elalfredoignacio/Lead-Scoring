## EDA CATEGÓRICAS

### Análisis estadístico

```python
def frecuencias_cat(df_cat):
    resultado = df_cat.apply(lambda x: x.value_counts(normalize = True)).T.stack()\
                .to_frame().reset_index()\
                .rename(columns={'level_0':'Variable','level_1':'Valor',0: "Frecuencia"})\
                .sort_values(by = ['Variable','Frecuencia'])
    return(resultado)
```

```python
frecuencias_cat(cat)
```

1. Elimino los registros con "no_llamar" con el valor "OTROS" (que en este caso es Sí). Recordar que estamos en un caso de gestión de Leads.
2. Elimino los registros "no_enviar_email" que sean "Yes", dado que estos clientes no quiere ser contactados por mail y por ende no voy a tener forma de llegar a ellos.
3. Elimino los registros "ult_actividad" igual a "Email Bounced" dado que estos clientes no tienen una dirección de correo válida.
4. Eliminar las variables "conociste_" dado que en todos ellos la categoría "No" es más del 99% y por ende no aportan valor
5. Eliminar las variables "no_llamar" y "no_enviar_email", porque de acuerdo a 1 y 2 quedaron con un único valor.

```python
cat = cat.loc[(cat.no_llamar != 'OTROS') & (cat.no_enviar_email != 'Yes') & (cat.ult_actividad != 'Email Bounced')] \
    .drop(columns = ['conociste_facebook', 'conociste_google', 'conociste_periodico', 'conociste_referencias',
                    'no_llamar', 'no_enviar_email'])
cat
```

IMPORTANTE: Eliminar los mismos registros en el dataset "num"

```python
num = num[num.index.isin(cat.index)]
num
```

### Análisis gráfico

```python
def graficos_eda_categoricos(cat):
    
    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(cat)
```

## EDA NUMÉRICAS

### Análisis estadístico

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(num)
```

### Análisis gráfico

```python
def graficos_eda_continuas(num):
    
    #Calculamos el número de fila que necesitamos
    from math import ceil
    filas = ceil(num.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(num):
        num[variable].plot.density(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_continuas(num)
```

Dado que todas las distribuciones se aproximan bastante a dist Normal dentro de lo que nos podemos encontrar en la realidad (incluyendo las multi-modales), a priori no seria necesario hacer una proceso de normalizacion posterior
