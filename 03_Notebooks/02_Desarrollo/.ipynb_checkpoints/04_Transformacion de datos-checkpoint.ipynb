## TRANSFORMACIÓN DE CATEGÓRICAS

Target Encoding es válida para las categóricas, pero por un tema de simplificación, voy a hacer OHE u OE.
OE es para las variables ordinales. Como mis variables categóricas no tiene un orden a priori, aplico a todas OHE.

### One Hot Encoding

#### Variables a aplicar OHE

```python
var_ohe = ['origen','fuente','ult_actividad','ambito', 'ocupacion', 'descarga_lm']
```

#### Instanciar

```python
ohe = OneHotEncoder(sparse = False, handle_unknown='ignore')
```

#### Entrenar y aplicar

```python
cat_ohe = ohe.fit_transform(cat[var_ohe])
```

#### Guardar como dataframe

```python
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out())
```

```python
cat_ohe
```

En el cat_ohe se perdió el id que era el índice. Lo re-establezco

```python
cat_ohe.set_index(cat.index, inplace = True)
```

## TRANSFORMACIÓN DE NUMÉRICAS

Como acá nuestro foco es la predicción y no la caracterización, no iremos por el camino de la DISCRETIZACIÓN ni la BINARIZACIÓN

La normalización como se vio en el EDA, no es necesario

## UNIFICAR DATASETS TRANSFORMADOS

### Unir todos los dataframes

```python
# OJO con los indices de los dataframes
df = pd.concat([cat_ohe, num], axis = 1)
```

```python
df
```

## REESCALAR VARIABLES

El reescalado es fundamental si queremos aplicar K-Means. No voy a usar Robust Scaler porque ya hice tratamiento de atípicos.

Por otro lado como voy a hacer K-Means y ya usé OHE, voy a usar Min-Max Scaling para las numéricas para que todas las variables (cat y num) me queden entre 0 y 1.

La variable "compra" (target) ya viene entre 0 y 1 así que no la vamos a transformar

### Con Min-Max

```python
df.info()
```

#### Variables a reescalar con Min-Max

Dejo afuera las variables codificadas en OHE

```python
var_mms = df.iloc[:,38:].columns
```

#### Instanciar

```python
mms = MinMaxScaler()
```

#### Entrenar y aplicar

```python
df_mms = mms.fit_transform(df[var_mms])
```

#### Guardar como dataframe

```python
#Añadir sufijos a los nombres
nombres_mms = [variable + '_mms' for variable in var_mms]

#Guardar como dataframe
df_mms = pd.DataFrame(df_mms,columns = nombres_mms)
```

```python
df_mms
```

Re-establezco el indice de df_mms con el id

```python
df_mms.set_index(df.index, inplace = True)
```

## UNIFICAR DATASETS REESCALADOS

### Crear una lista con los dataframes a incluir en el tablón analítico

```python
incluir = [df_mms, cat_ohe, target]
```

### Unir todos los dataframes en el tablón analítico

```python
df_tablon = pd.concat(incluir, axis = 1)
```
